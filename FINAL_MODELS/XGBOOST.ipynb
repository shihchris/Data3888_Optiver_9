```python
import os, glob
import pandas as pd

# Paths

output_dir = 'splits'   # ← this will create ./splits/

# 1. Load
train_df = df_30s.copy()
test_df  = df_30s_test.copy()

# 3. Split & write
for df, kind in [(train_df, 'train'), (test_df, 'test')]:
    for (sid, wk), grp in df.groupby(['stock_id','week'], sort=False):
        if kind=='train':
            grp_sub = grp[grp['day_of_week'] < 5]
        else:
            grp_sub = grp[grp['day_of_week'] == 5]
        if grp_sub.empty: continue
        folder = os.path.join(output_dir, str(sid), f"week_{wk}")
        os.makedirs(folder, exist_ok=True)
        grp_sub.reset_index(drop=True).to_parquet(
            os.path.join(folder, f"{kind}.parquet"), index=False
        )

# 4. Verify one example
dirs = sorted(glob.glob(os.path.join(output_dir, '*', 'week_*')))
print("Total groups:", len(dirs))
if dirs:
    example = dirs[0]
    print("Example:", example, os.listdir(example))
    print("Train preview:")
    print(pd.read_parquet(os.path.join(example,'train.parquet')).head())
    print("Test preview:")
    print(pd.read_parquet(os.path.join(example,'test.parquet')).head())
```

```python
# =============================================================
# Incremental XGBoost: static fit → 10‑min warm‑start updates
# =============================================================
import numpy as np, pandas as pd, matplotlib.pyplot as plt, math
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# ─── 1. CONFIG ────────────────────────────────────────────────
FEATURES = ["abs_volatility","spread","imbalance",
            "depth","depth_ratio","rv_rolling","mom_rolling"]
TARGET   = "realized_volatility"
SCALE    = 10_000                       # scale the target
CHUNK_S  = 300                          # 10‑min blocks
ADDITIONAL_TREES = 40                   # trees added each 10‑min update
UPDATE_LR = 0.01          # << small
ADD_TREES = 20


base_params = dict(
    n_estimators = 410,                 # initial trees
    max_depth    = 5,
    learning_rate= 0.13698888595774816,
    subsample    = 0.6021495265369203,
    colsample_bytree = 0.7350403798567642,
    gamma        = 0.0,                 # splits allowed
    reg_alpha    = 0.0030978353003809256,
    reg_lambda   = 0.09512970484709227,
    objective    = "reg:squarederror",
    random_state = 42,
    verbosity    = 0
)

# ─── 2. PREPARE TRAIN & TEST ──────────────────────────────────
train = df_train_8382_w1.copy().reset_index(drop=True)
test  = df_test_8382_w1 .copy().reset_index(drop=True)

train_y = train[TARGET] * SCALE
test_y  = test [TARGET] * SCALE                        # keep scaled for later metrics

scaler = StandardScaler().fit(train[FEATURES])
train_X = scaler.transform(train[FEATURES])
test_X  = scaler.transform(test [FEATURES])

test_raw_sec   = test["time_seconds"].copy()           # keep wall‑clock seconds
test_chunks    = (test_raw_sec // CHUNK_S).astype(int) # 10‑min IDs
test["chunk"]  = test_chunks

# ─── 3. INITIAL BASE MODEL ───────────────────────────────────
model = XGBRegressor(**base_params)
model.fit(train_X, train_y)

# store everything seen so far
hist_X = train_X.copy()
hist_y = train_y.copy()

# collectors
preds_scaled, actuals_scaled, times_sec = [], [], []

# ─── STREAM WITH SMALL INCREMENTS ─────────────────────────────
history_X, history_y = train_X.copy(), train_y.copy()
times_sec, actuals_scaled, preds_scaled = [], [], []

# base model already fitted above
for _, chunk in test.groupby("chunk", sort=True):
    cX = test_X[chunk.index]
    cy = test_y.iloc[chunk.index].values

    # (a) predict current block
    p_scaled = model.predict(cX)
    preds_scaled.extend(p_scaled)
    actuals_scaled.extend(cy)
    times_sec.extend(test_raw_sec.loc[chunk.index])

    # (b) grow a **small** stage of trees
    incr_params = base_params.copy()
    incr_params.update(n_estimators=ADD_TREES, learning_rate=UPDATE_LR)

    model.n_estimators += ADD_TREES
    model.set_params(**incr_params)
    history_X = np.vstack([history_X, cX])
    history_y = np.concatenate([history_y, cy])
    model.fit(history_X, history_y, xgb_model=model)

# ─── 5. DAY‑5 (11:00‑17:00) VISUAL & METRICS ──────────────────
# identify Day‑5 rows (index 4 because Monday=0)
day_idx   = np.array(times_sec)//86400
mask_day5 = day_idx == 4

# hour‑of‑day
hod = (np.array(times_sec)[mask_day5] % 86400)/3600.0
session_start = 11.0               # we know trading begins here
session_end   = 17.0

# convert back to real units
act5  = np.array(actuals_scaled)[mask_day5]/SCALE
pred5 = np.array(preds_scaled)[mask_day5]/SCALE

fig, ax = plt.subplots(figsize=(12,5))
ax.plot(hod, act5,  label="Actual",    alpha=0.8)
ax.plot(hod, pred5, "--", label="Predicted", alpha=0.8)

for h in range(int(session_start), int(session_end)+1):
    ax.axvline(h, color="k", linestyle="--", alpha=0.3)

ax.set_xlim(session_start, session_end)
ax.set_xticks(range(int(session_start), int(session_end)+1))
ax.set_xticklabels([f"{h}:00" for h in range(int(session_start), int(session_end)+1)])

ax.set_xlabel("Hour of day")
ax.set_ylabel("Realized volatility")
ax.set_title("Rolling 10‑Minute Warm‑start XGB — Stock 8382.0, Day 5 (11‑17)")
ax.legend()
plt.tight_layout()
plt.show()

# metrics
r2   = r2_score(act5, pred5)
mse  = mean_squared_error(act5, pred5)
mae  = mean_absolute_error(act5, pred5)
rmse = math.sqrt(mse)

print("Day 5 performance (11‑17, warm‑start XGB):")
print(f"  R²   : {r2: .4f}")
print(f"  MSE  : {mse: .6e}")
print(f"  MAE  : {mae: .6e}")
print(f"  RMSE : {rmse: .6e}")
```

```python
# =============================================================================
# Robust XGBoost Volatility Forecasting
# =============================================================================
import warnings, numpy as np, pandas as pd, matplotlib.pyplot as plt
warnings.filterwarnings("ignore")

import xgboost as xgb
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    mean_squared_error, mean_absolute_error,
    median_absolute_error, r2_score,
    mean_absolute_percentage_error
)
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from scipy.stats import randint, uniform, loguniform

# ───────── USER CONFIG ───────────────────────────────────────
FEATURE_COLS  = ["abs_volatility","spread","imbalance",
                 "depth","depth_ratio","rv_rolling","mom_rolling"]
LAG_STEP      = 1                 # one 5-min block lag
TARGET_COL    = "realized_volatility"
TIME_COL      = "time_seconds"
DAY_COL       = "day_of_week"     # 1=Mon … 5=Fri
CHUNK_SEC     = 300               # 5-min block
RETRAIN_SEC   = 1800              # 30-min retrain
SCALE_TARGET  = 10_000
N_RANDOM      = 40
RANDOM_SEED   = 42

# ───────── LOAD & PREP DATA ─────────────────────────────────
train_df = df_30s.copy()
test_df  = df_30s_test.copy()
full_df  = pd.concat([train_df, test_df]) \
             .sort_values(["stock_id","week",TIME_COL]) \
             .reset_index(drop=True)

# 1) lag features
def lag_features(df, groups, cols, lag=1):
    return df.assign(**{
        f"{c}_lag": df.groupby(groups)[c].shift(lag)
        for c in cols
    })

full_df = lag_features(full_df, ["stock_id","week"], FEATURE_COLS, lag=LAG_STEP)
full_df.dropna(subset=[f"{c}_lag" for c in FEATURE_COLS], inplace=True)
LAG_FEATURES = [f"{c}_lag" for c in FEATURE_COLS]

# 2) scale
scaler = StandardScaler().fit(full_df.loc[full_df[DAY_COL]<5, LAG_FEATURES])
full_df[LAG_FEATURES] = scaler.transform(full_df[LAG_FEATURES])

# ───────── RANDOM SEARCH TUNER ───────────────────────────────
def tune_xgb_random(X, y):
    tscv = TimeSeriesSplit(n_splits=4)
    param_dist = {
        "max_depth":        randint(3, 11),
        "learning_rate":    loguniform(1e-3, 3e-1),
        "n_estimators":     randint(80, 400),
        "subsample":        uniform(0.5, 0.5),
        "colsample_bytree": uniform(0.5, 0.5),
        "gamma":            loguniform(1e-8, 1.0),
        "min_child_weight": randint(1, 11)
    }
    base = xgb.XGBRegressor(
        objective="reg:squarederror",
        tree_method="hist",
        random_state=RANDOM_SEED
    )
    rs = RandomizedSearchCV(
        base, param_dist, n_iter=N_RANDOM, cv=tscv,
        scoring="neg_root_mean_squared_error",
        random_state=RANDOM_SEED, n_jobs=-1
    )
    rs.fit(X, y)
    best = rs.best_params_
    best.update(objective="reg:squarederror",
                tree_method="hist",
                random_state=RANDOM_SEED)
    return best

# ───────── METRICS FUNCTION ─────────────────────────────────
def compute_vol_metrics(actuals, preds):
    rmse  = mean_squared_error(actuals, preds, squared=False)
    mae   = mean_absolute_error(actuals, preds)
    medae = median_absolute_error(actuals, preds)
    r2    = r2_score(actuals, preds)
    mape  = mean_absolute_percentage_error(actuals, preds) * 100
    smape = np.mean(
        2 * np.abs(preds - actuals) /
        (np.abs(preds) + np.abs(actuals) + 1e-12)
    ) * 100
    qlike = np.mean(np.log(preds**2) + (actuals**2) / (preds**2))
    return {
        "RMSE": rmse / SCALE_TARGET,
        "MAE": mae / SCALE_TARGET,
        "MedAE": medae / SCALE_TARGET,
        "MAPE(%)": mape,
        "SMAPE(%)": smape,
        "QLIKE": qlike,
        "R2": r2
    }

# ───────── 4. ROLLING FORECAST ───────────────────────────────
def rolling_forecast(X_base, y_base, X_fri, y_fri, params, step_rows):
    model = xgb.XGBRegressor(**params).fit(X_base, y_base)
    preds_xgb, preds_naive, actuals, idxs = [], [], [], []
    last_rv = y_base.iloc[-1]
    X_hist, y_hist = X_base.copy(), y_base.copy()

    for st in range(0, len(X_fri), step_rows):
        ed = min(st+step_rows, len(X_fri))
        X_blk, y_blk = X_fri.iloc[st:ed], y_fri.iloc[st:ed]

        preds_xgb.extend(model.predict(X_blk))
        preds_naive.extend([last_rv] * len(X_blk))
        actuals.extend(y_blk)
        idxs.extend(X_blk.index)

        last_rv = y_blk.iloc[-1]
        X_hist = pd.concat([X_hist, X_blk])
        y_hist = pd.concat([y_hist, y_blk])
        if ed < len(X_fri):
            model = xgb.XGBRegressor(**params).fit(X_hist, y_hist)

    return np.array(preds_xgb), np.array(preds_naive), np.array(actuals), idxs

# ───────── 5. MAIN LOOP ─────────────────────────────────────
metrics = []
first_plot = True
step_rows = int(RETRAIN_SEC / (full_df[TIME_COL].diff().median() or CHUNK_SEC))

for (sid, wk), grp in full_df.groupby(["stock_id","week"]):
    mon_thu = grp[grp[DAY_COL] < 5]
    fri     = grp[grp[DAY_COL] == 5]
    if fri.empty or mon_thu.empty:
        continue

    X_train = mon_thu[LAG_FEATURES]
    y_train = mon_thu[TARGET_COL] * SCALE_TARGET
    X_fri   = fri[LAG_FEATURES]
    y_fri   = fri[TARGET_COL] * SCALE_TARGET

    best = tune_xgb_random(X_train, y_train)
    pred_xgb, pred_naive, act, idx = rolling_forecast(
        X_train, y_train, X_fri, y_fri, best, step_rows)

    # compute metrics
    m_x = compute_vol_metrics(act, pred_xgb)
    m_n = compute_vol_metrics(act, pred_naive)

    # record both
    m_x.update({"stock": sid, "week": wk, "model": "XGB"})
    m_n.update({"stock": sid, "week": wk, "model": "Naive"})
    metrics.append(m_x)
    metrics.append(m_n)

    # plot first group
    if first_plot:
        hod = (fri.loc[idx, TIME_COL] % 86400)/3600
        plt.figure(figsize=(12,4))
        plt.plot(hod, act/SCALE_TARGET, label="Actual")
        plt.plot(hod, pred_xgb/SCALE_TARGET, '--', label="XGB")
        plt.plot(hod, pred_naive/SCALE_TARGET, ':', label="Naive")
        plt.vlines(range(11,18), *plt.ylim(), ls='--', alpha=.3)
        plt.title(f"Stock {sid} Week {wk}")
        plt.xlabel("Hour"); plt.ylabel("Realized Volatility")
        plt.legend(); plt.tight_layout(); plt.show()
        first_plot = False

    print(f"[{sid}-{wk}] XGB R²={m_x['R2']:.3f} vs Naive R²={m_n['R2']:.3f}")

# ───────── 6. RESULTS TABLE ─────────────────────────────────
met_df = pd.DataFrame(metrics)
fmt = {
    'R2': "{:.3f}".format,
    'RMSE': "{:.2e}".format,
    'MAE': "{:.2e}".format,
    'MedAE': "{:.2e}".format,
    'MAPE(%)': "{:.2f}%".format,
    'SMAPE(%)': "{:.2f}%".format,
    'QLIKE': "{:.3f}".format
}
print("\n=== Metrics (Mon–Thu → Fri, 30-min expanding) ===")
print(met_df.to_string(index=False, formatters=fmt))
```
